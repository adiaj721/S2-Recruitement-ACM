```python
df.isnull().sum()
```

```python
#if total charges appear as object they are not detected as null hence we convert
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
#now we can see that the totalcharges has changed
#TotalCharges was converted to numeric. Invalid entries were coerced into NaN for proper missing value handling.
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
df.isnull().sum()
```

```python
#Median imputation was chosen for TotalCharges because the feature is skewed and contains outliers, making median more robust than mean.
```

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)
```

```python
#Outliers were treated using IQR-based capping instead of removal to avoid losing valuable customer information.
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.describe()
```

```python
#Duplicate records were checked and removed to ensure data integrity.
#no duplicates found
```

```python
numeric_cols = ["tenure", "MonthlyCharges", "TotalCharges"]
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
#binary encoding
```

```python
df = pd.get_dummies(df, drop_first=True)
#Binary variables were label-encoded, while multi-category features were one-hot encoded to avoid ordinal assumptions.
```

```python
#TotalCharges may introduce data leakage as it summarizes customer payment history which could indirectly encode churn behavior.
#Additionally, performing scaling or encoding before train-test splitting can leak test-set information into training.
```
