```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.columns
```

```python
df['ExamScore'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
#we are taking the mean of exam scores of 3 subjects into single
```

```python
#now lets plot each  graph
#first is maths exam graph
```

```python
sns.histplot(df['math score'], kde=True)
plt.title("Distribution of Overall Exam Score")
plt.show()
#this is distribution of overall student perfomance for maths
```

```python
#The math scores appear to be roughly normally distributed, with most students scoring between 60 and 80.
```

```python
#2nd graph is reading vs writing scores distribution
```

```python
plt.figure()
sns.scatterplot(x=df["reading score"], y=df["writing score"])
plt.title("Reading Score vs Writing Score")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```

```python
The points form a clear upward pattern, showing that students who score higher in reading also tend to score higher in writing.
```

```python
#3rd plot
```

```python
sns.boxplot(x=df['test preparation course'], y=df['reading score'])
plt.title(" Reading Score by test preparation course")
plt.show()
```

```python
Students who completed the test preparation course generally
show higher reading scores compared to those who did not.
The difference in medians suggests that test preparation may
had a positive impact on reading performance.
```

```python
#4th gender vs exam scores
sns.boxplot(x=df['gender'], y=df['ExamScore'])
plt.title("Overall Exam Score by Gender")
plt.show()
```

```python
#5th
sns.boxplot(x=df['lunch'], y=df['ExamScore'])
plt.title("Exam Score by Lunch Type")
plt.show()
```

```python
sns.heatmap(
    df[['math score', 'reading score', 'writing score']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Between Subject Scores")
plt.show()
```

```python
#now lets analyse the skewness
#Skewness analysis helps identify whether score distributions are symmetric or skewed, which may impact further modeling.
```

```python
df[['math score', 'reading score', 'writing score', 'ExamScore']].skew()
```
